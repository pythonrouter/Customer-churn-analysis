```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('customer churn.csv')
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
         return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
#converted 0 and 1 values of senior citizen to yes/no to make it easier to understand
```

```python
ax = sns.countplot(x ='Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("count of customers by churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby('Churn').agg({'Churn' : "count"})
gb
plt.pie(gb['Churn'],labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers")
plt.show
```

```python
# from the given pie chart we can conclude that 26.54% of our customer have churned out
```

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.show()
```

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "SeniorCitizen", data = df, hue = "Churn")
plt.show()
```

```python
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='count')
total = grouped.groupby('SeniorCitizen')['count'].transform('sum')
grouped['percentage'] = (grouped['count'] / total * 100)

# Pivot table for plotting
pivot_table = grouped.pivot(index='SeniorCitizen', columns='Churn', values='percentage').fillna(0)

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(4, 4))  # Adjust figure size if needed
categories = pivot_table.index
bottom = [0] * len(categories)  # Initialize bottom values for stacking

# Plot each segment
for churn_status in pivot_table.columns:
    ax.bar(categories, pivot_table[churn_status], bottom=bottom, label=churn_status)
    # Annotate each bar segment
    for i, percentage in enumerate(pivot_table[churn_status]):
        if percentage > 0:
            ax.text(i, bottom[i] + percentage / 2, f"{percentage:.1f}%", ha='center', va='center', color='white')
    bottom = [b + p for b, p in zip(bottom, pivot_table[churn_status])]

# Customize the plot
ax.set_title("Churn Percentage by Senior Citizen Status")
ax.set_xlabel("Senior Citizen (0=No, 1=Yes)")
ax.set_ylabel("Percentage")
ax.legend(title="Churn", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()













```

```python
# comparatively a greater percentage of people in senior category have churned out
```

```python
# Columns for which we want count plots
columns = [
    'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
    'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
    'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod',
    'MonthlyCharges', 'TotalCharges', 'Churn'
]

# Define the number of rows and columns for the subplots
n_cols = 3  # Number of plots per row
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate required rows

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))

# Flatten axes for easier iteration if more than one row
axes = axes.flatten()

# Loop through columns and create count plots
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i] , hue = df["Churn"])
    axes[i].set_title(f'Count Plot for {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

# Remove any unused subplots
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#This data visualization shows count plots for various categorical variables and their relationship with customer churn (e.g., "Yes" or "No"). Key trends are evident, such as higher churn rates for customers with certain
#Internet services or payment methods
```
